```python
class VGG16(nn.modules):

    def __init__(self):
        super(VGG16, self).init()
        self.layer1  =  nn.Sequential(
        nn.Conv2d(in_channels=3, out_channels=64, kernel_size=3, stride=1, padding=1),
        relu(inplace=True),        
        )
        self.layer2 = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=64, stride=1, padding=1),
            relu(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2, padding=0, dilation=1, ceil_mode=False)
        )
        self.layer3  = nn.Sequential(
            nn.Conv2d(in_channels=64, out_channels=128, kernel_size=3, stride=1, padding=1),
            relu(inplace=True)
        )

        self.layerr4 = nn.Sequential(
            nn.Conv2d(in_channels=128, out_channels=128, kernel_size=3, stride=1 ,padding=1),
            relu(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2, padding=0)        
        )

        self.layer5 = nn.Sequential(
            nn.Conv2d(in_channels=128, out_channels=256, kernel_size=3, stride=1, padding=1),
            relu(inplace=True)                 
                                    )
        self.layer6 = nn.Sequential(
            nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, stride=1, padding=1),
            relu(inplace=True)
                                    )
        self.layer7 = nn.Sequential(
            nn.Conv2d(in_channels=256, out_channels=256, kernel_size=3, stride=1, padding=1),
            relu(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
    
        )
        self.layer8 = nn.Sequential(
            nn.Conv2d(in_channels=256,out_channels=512, kernel_size=3, stride=1, padding=1),
            relu(inplace=True)
        )

        self.layer9 = nn.Sequential(
            nn.Conv2d(in_channels= 512,out_channels=512, kernel_size=3, stride=1, padding=1),
            relu(inplace=True)
        )

        self.layer10 = nn.Sequential(
            nn.Conv2d(
                in_channels=512, out_channels=512, kernel_size=3, stride=1, padding=1
            ),
            relu(inplace=True),
        )

        self.layer11 = nn.Sequential(
            nn.Conv2d(in_channels=512,out_channels= 512, kernel_size=3, stride=1, padding=1),
            relu(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        )

        self.layer12 = nn.Sequential(
            nn.Conv2d(
                in_channels=512, out_channels=512, kernel_size=3, stride=1, padding=1
            ),
            relu(inplace=True),
        )
        self.layer13 = nn.Sequential(
            nn.Conv2d(
                in_channels=512, out_channels=512, kernel_size=3, stride=1, padding=1
            ),
            relu(inplace=True),
        )
        self.layer14 = nn.Sequential(
            nn.Conv2d(in_channels=512,out_channels= 512, kernel_size=3, stride=1, padding=1),
            relu(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2, padding=0)
        )

        self.layer15 = nn.AdaptiveAvgPool2d(output_size=(7, 7))

        self.layer16 = nn.Sequential(
            nn.LazyLinear(out_features=4096, bias=True),
            relu(inplace=True),
            nn.Dropout(p=0.5, inplace=False),
            nn.Linear(in_features=4096 ,out_features=1000, bias=True),
            relu(inplace = True),         
            nn.Dropout(p=0.5, inplace=False),
            nn.Linear(in_features=1000, out_features=10, bias=True)  
    
        )

        def forward(self, x):
            out = self.layer1(x)
            out = self.layer2(out)
            out = self.layer3(out)
            out = self.layer4(out)
            out = self.layer5(out)
            out = self.layer6(out)
            out = self.layer7(out)
            out = self.layer8(out)
            out = self.layer9(out)
            out = self.layer10(out)
            out = self.layer11(out)
            out = self.layer12(out)
            out = self.layer13(out)
            out = self.layer14(out)
            out = self.layer15(out)
            out = out.view(out.size(0), -1)
            out = self.layer16(out)
            
            return out

```
